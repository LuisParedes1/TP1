Obtenga la matriz de distancias euclídeas para todos los contenidos que están en Marte. ¿Cuáles son los dos contenidos que están a menor distancia?  

```python
# importacion general de librerias y de visualizacion (matplotlib y seaborn)
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.feature_extraction.text import CountVectorizer
from sklearn.neighbors import NearestNeighbors
from sklearn.metrics.pairwise import euclidean_distances 

%matplotlib inline

plt.style.use('default') # haciendo los graficos un poco mas bonitos en matplotlib
#plt.rcParams['figure.figsize'] = (20, 10)

sns.set(style="whitegrid") # seteando tipo de grid en seaborn

pd.options.display.float_format = '{:20,.2f}'.format # suprimimos la notacion cientifica en los outputs

import warnings
warnings.filterwarnings('ignore')

from datetime import datetime
```

```python
geo_tags_db = pd.read_csv('./Wikipedia_dataset_2c2021/geo_tags.csv')

contenidos = pd.read_csv('./Wikipedia_dataset_2c2021/contents_text_sample.csv')
```

# 1. Busco los contenidos que estan en marte

```python
contenido_marte = geo_tags_db[geo_tags_db['gt_globe']=='mars']
display(contenido_marte)
```

```python
contenido_interes = contenidos.merge(contenido_marte, how='inner', left_on='id', right_on='gt_page_id' )
display(contenido_interes)
```

# 2. Limpio los textos

```python
contenido_interes['texto_limpio'] = contenido_interes['text'].map(lambda x: str(x).replace('\n', '')
         .replace('.', '')
         .replace('|','')
         .replace('[','')
         .replace(']','')
         .replace('{','')
        .replace('}','')
)
```

```python
corpus = contenido_interes['texto_limpio'].to_list()

vectorizer = CountVectorizer()

matriz_embedding = vectorizer.fit_transform(corpus)
```

```python
matriz_euclediana = euclidean_distances(matriz_embedding)

pd.DataFrame(matriz_euclediana)
```

# 3. Hallo la mínima distancia

```python
min_distancia = matriz_euclediana[matriz_euclediana>0].min()
display(min_distancia)
```

# 4. Hallo los textos correspondientes mas cercanos 

```python
np.where(matriz_euclediana==min_distancia)
```
