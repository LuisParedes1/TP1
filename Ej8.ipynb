La mediana de la antigüedad para las últimas ediciones vigentes agrupado por el primer carácter del título

```python
# importacion general de librerias y de visualizacion (matplotlib y seaborn)
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import pytz
from datetime import datetime, timedelta

%matplotlib inline

plt.style.use('default') # haciendo los graficos un poco mas bonitos en matplotlib
#plt.rcParams['figure.figsize'] = (20, 10)

sns.set(style="whitegrid") # seteando tipo de grid en seaborn

pd.options.display.float_format = '{:20,.2f}'.format # suprimimos la notacion cientifica en los outputs

import warnings
warnings.filterwarnings('ignore')

from datetime import datetime
```

```python
# Cargo los datasets
wiki_contents=pd.read_csv('./Wikipedia_dataset_2c2021/contents.csv', usecols=['title', 'revision_timestamp'])
```

```python
wiki_contents
```

## 1. Convierto las fechas de ultima revision en formato datetime

```python
wiki_contents['revision_timestamp'] = pd.to_datetime(wiki_contents['revision_timestamp'])
```

# 2. Calculo la antiguedad para cada articulo

```python
wiki_contents['antiguedad'] = wiki_contents['revision_timestamp'].map( lambda x: datetime.now(pytz.utc) - x )
display(wiki_contents)
```

# 3. Agrupo por primera letra del titulo y aplico la mediana de la antiguedad

```python
wiki_contents['inicial_tit'] = wiki_contents['title'].transform(lambda x: str(x)[0])
display(wiki_contents)
```

```python
wiki_contents.groupby('inicial_tit').agg(antiguedad=('antiguedad', 'median') , muestra = ('antiguedad', 'count'))
```

# Conclusion

Se obtuvo la mediana de la antigüedad de las ultimas ediciones agrupadas por su primera letra del titulo. 

Es claro que para algunas letras o caracteres particulares como 𠘨, 🧆, 𠆢 se tiene una muestra muy pequeña puesto que posiblemente representa el articulo de la descripcion del mismo y no hay mas articulos que empiecen con esos caracteres en el titulo. 

Para estos casos la formula mas peligrosa del mundo no les afecta puesto que se tiene una muestra con datos de todos los contenidos de Wikipedia en su versión más reciente. Es decir, tenemos una muestra representativa del total, por ende esta mediana es correcta.
